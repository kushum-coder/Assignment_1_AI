#"Using conventional thresholds to categorize nations into HDI groups."

```python
#6 Adding HDI Category Column
def hdi_category(val):
    if val < 0.550:
        return "Low"
    elif val < 0.700:
        return "Medium"
    elif val < 0.800:
        return "High"
    else:
        return "Very High"

hdi_2022_df["HDI Category"] = hdi_2022_df["hdi"].apply(hdi_category)
hdi_2022_df
```

#"Using statistical techniques to identifying extreme development values."

```python
# ---- 7: Outlier Detection for HDI and GNI per Capita ----

# Ensure sa_df exists (run previous cell if not)
# Function to detect outliers using 1.5 * IQR rule
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[column] < Q1 - 1.5*IQR) | (df[column] > Q3 + 1.5*IQR)]
    return outliers

# Detect outliers
hdi_outliers = detect_outliers_iqr(sa_df, "hdi")
gni_outliers = detect_outliers_iqr(sa_df, "gross_inc_percap")

# Scatter plot with HDI outliers highlighted
plt.figure(figsize=(8,6))
sns.scatterplot(data=sa_df, x="gross_inc_percap", y="hdi", label="Normal")
sns.scatterplot(data=hdi_outliers, x="gross_inc_percap", y="hdi", color='red', label='HDI Outlier')
plt.title("HDI vs GNI per Capita with Outliers (South Asia)")
plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.legend()
plt.show()
```

#"Assessing the relationship between development indicators and HDI."

```python
# ---- 8: Correlation Analysis ----

metrics = ["life_expectancy", "gender_development"]

for metric in metrics:
    corr = sa_df[metric].corr(sa_df["hdi"])
    print(f"Pearson correlation between HDI and {metric}: {corr:.3f}")

    plt.figure(figsize=(7,5))
    sns.scatterplot(data=sa_df, x=metric, y="hdi")
    sns.regplot(data=sa_df, x=metric, y="hdi", scatter=False, color='red')
    plt.title(f"HDI vs {metric} (South Asia)")
    plt.xlabel(metric)
    plt.ylabel("HDI")
    plt.show()
```

```python
# ---- 9: Composite Score vs HDI Ranking ----

# Create HDI and Composite rankings
sa_df["HDI Rank"] = sa_df["hdi"].rank(ascending=False)
sa_df["Composite Rank"] = sa_df["Composite Score"].rank(ascending=False)

# Display comparison table
comparison = sa_df[["country", "hdi", "HDI Rank", "Composite Score", "Composite Rank"]].sort_values("Composite Rank")
comparison
```

```python
# ---- Import Libraries ----
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ---- Load Dataset ----
df = pd.read_csv('/content/drive/MyDrive/Human_Development_Index_Dataset.csv', encoding='ISO-8859-1')
df.head()
```

#"Examining the differences in human development between the Middle East and South Asia."

```python
# ---- Problem 3 Cell 1: Subsets for South Asia and Middle East ----

# Define country groups
south_asia = [
    "Afghanistan", "Bangladesh", "Bhutan", "India",
    "Maldives", "Nepal", "Pakistan", "Sri Lanka"
]

middle_east = [
    "Bahrain", "Iran", "Iraq", "Israel", "Jordan",
    "Kuwait", "Lebanon", "Oman", "Palestine", "Qatar",
    "Saudi Arabia", "Syria", "United Arab Emirates", "Yemen"
]

# Filter 2020–2022 dataset
sa_2020_2022 = df_1b[df_1b["country"].isin(south_asia)].copy()
me_2020_2022 = df_1b[df_1b["country"].isin(middle_east)].copy()

# Optional: save CSVs for submission
sa_2020_2022.to_csv("HDI_SouthAsia_2020_2022.csv", index=False)
me_2020_2022.to_csv("HDI_MiddleEast_2020_2022.csv", index=False)

print("South Asia and Middle East subsets created.")
```

#"Providing an overview of regional variations in HDI levels."

```python
# ---- Problem 3 Cell 2: Descriptive Statistics ----

sa_mean, sa_std = sa_2020_2022["hdi"].mean(), sa_2020_2022["hdi"].std()
me_mean, me_std = me_2020_2022["hdi"].mean(), me_2020_2022["hdi"].std()

print("South Asia HDI: Mean =", round(sa_mean,3), ", Std =", round(sa_std,3))
print("Middle East HDI: Mean =", round(me_mean,3), ", Std =", round(me_std,3))
```

#"Determining which nations in each region are performing the best and worst."

```python
# ---- Problem 3 Cell 3: Top and Bottom HDI Countries ----

# South Asia
sa_top3 = sa_2020_2022.groupby("country")["hdi"].mean().sort_values(ascending=False).head(3)
sa_bottom3 = sa_2020_2022.groupby("country")["hdi"].mean().sort_values().head(3)

# Middle East
me_top3 = me_2020_2022.groupby("country")["hdi"].mean().sort_values(ascending=False).head(3)
me_bottom3 = me_2020_2022.groupby("country")["hdi"].mean().sort_values().head(3)

print("South Asia Top 3 Countries:\n", sa_top3)
print("\nSouth Asia Bottom 3 Countries:\n", sa_bottom3)
print("\nMiddle East Top 3 Countries:\n", me_top3)
print("\nMiddle East Bottom 3 Countries:\n", me_bottom3)

# Optional: Bar plot for top/bottom performers
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.barplot(x=sa_top3.values, y=sa_top3.index, color='blue', label='SA Top')
sns.barplot(x=sa_bottom3.values, y=sa_bottom3.index, color='cyan', label='SA Bottom')
sns.barplot(x=me_top3.values, y=me_top3.index, color='red', label='ME Top')
sns.barplot(x=me_bottom3.values, y=me_bottom3.index, color='pink', label='ME Bottom')
plt.xlabel("Average HDI (2020–2022)")
plt.title("Top & Bottom HDI Countries: South Asia vs Middle East")
plt.legend()
plt.show()
```

#"Examining important developmenet metrics by area."

```python
# ---- Problem 3 Cell 4: Metric Comparisons ----

metrics = ["gender_development", "life_expectancy", "gross_inc_percap"]

# Compute region averages
sa_metrics = sa_2020_2022.groupby("year")[metrics].mean().reset_index()
me_metrics = me_2020_2022.groupby("year")[metrics].mean().reset_index()

# Plot each metric
for metric in metrics:
    plt.figure(figsize=(8,5))
    sns.barplot(x="year", y=metric, data=sa_metrics, color='blue', alpha=0.6, label='South Asia')
    sns.barplot(x="year", y=metric, data=me_metrics, color='red', alpha=0.6, label='Middle East')
    plt.title(f"Comparison of {metric} (2020–2022)")
    plt.ylabel(metric)
    plt.legend()
    plt.show()
```

#"Using statistical measures to evaluate development drivers and inequalities."

```python
# ---- Problem 3 Cell 5: HDI Disparity ----

# Range
sa_range = sa_2020_2022["hdi"].max() - sa_2020_2022["hdi"].min()
me_range = me_2020_2022["hdi"].max() - me_2020_2022["hdi"].min()

# Coefficient of Variation (CV = std/mean)
sa_cv = sa_2020_2022["hdi"].std() / sa_2020_2022["hdi"].mean()
me_cv = me_2020_2022["hdi"].std() / me_2020_2022["hdi"].mean()

print("South Asia HDI: Range =", round(sa_range,3), ", CV =", round(sa_cv,3))
print("Middle East HDI: Range =", round(me_range,3), ", CV =", round(me_cv,3))
```

```python
# ---- Problem 3 Cell 6: Correlation of HDI with metrics ----

# Compute correlations
sa_corr = sa_2020_2022[["hdi","gender_development","life_expectancy"]].corr()["hdi"]
me_corr = me_2020_2022[["hdi","gender_development","life_expectancy"]].corr()["hdi"]

print("South Asia Correlations with HDI:\n", sa_corr)
print("\nMiddle East Correlations with HDI:\n", me_corr)

# Scatter plots
for metric in ["gender_development", "life_expectancy"]:
    plt.figure(figsize=(6,4))
    sns.scatterplot(data=sa_2020_2022, x=metric, y="hdi", color='blue', label='South Asia')
    sns.regplot(data=sa_2020_2022, x=metric, y="hdi", scatter=False, color='navy')
    sns.scatterplot(data=me_2020_2022, x=metric, y="hdi", color='red', label='Middle East')
    sns.regplot(data=me_2020_2022, x=metric, y="hdi", scatter=False, color='darkred')
    plt.title(f"HDI vs {metric} (2020–2022)")
    plt.xlabel(metric)
    plt.ylabel("HDI")
    plt.legend()
    plt.show()
```

#"Showing development patterns, locate and display HDI outliers."

```python
# ---- Problem 3 Cell 7: Outlier Detection ----

def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    return df[(df[column] < Q1 - 1.5*IQR) | (df[column] > Q3 + 1.5*IQR)]

# Detect outliers
sa_hdi_outliers = detect_outliers_iqr(sa_2020_2022, "hdi")
me_hdi_outliers = detect_outliers_iqr(me_2020_2022, "hdi")

# Scatter plot with outliers
plt.figure(figsize=(8,6))
sns.scatterplot(data=sa_2020_2022, x="gross_inc_percap", y="hdi", label="SA Normal")
sns.scatterplot(data=sa_hdi_outliers, x="gross_inc_percap", y="hdi", color='blue', s=100, label='SA HDI Outlier')
sns.scatterplot(data=me_2020_2022, x="gross_inc_percap", y="hdi", label="ME Normal")
sns.scatterplot(data=me_hdi_outliers, x="gross_inc_percap", y="hdi", color='red', s=100, label='ME HDI Outlier')
plt.title("HDI vs GNI per Capita with Outliers")
plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.legend()
plt.show()
```
